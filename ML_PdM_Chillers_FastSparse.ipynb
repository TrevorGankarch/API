```python
# FastSparse_GAM_PdM.ipynb-style Python script
# ============================================================
# FastSparse-like GAM pipeline for Predictive Maintenance (PdM)
# Ready-to-run in a Jupyter notebook cell-by-cell.
# This notebook attempts to use a "FastSparse" implementation if available;
# otherwise it falls back to pyGAM + sparse feature selection.
# ============================================================

"""
Notebook overview (cells):

1) Setup: installs & imports
2) Load dataset (CSV) & preview
3) Feature engineering: lags, rolling stats
4) Train/val split
5) Sparse feature selection (L1 or tree-based)
6) Fit GAM (pyGAM) on selected features
7) Evaluate: classification / regression metrics
8) Interpret: partial dependence plots, GAM term plots
9) Save model & diagrams
10) Appendix: notes & references

Notes:
- Target variable: binary fault (0/1) or RUL (regression). The code auto-detects.
- If you specifically have a FastSparse implementation and it is pip-installable,
  change the import section to use it. I included a detection mechanism.

"""

# ---------------------------
# CELL 1 — Setup (run once)
# ---------------------------

# If running in a fresh environment, uncomment the pip installs below.
# Note: Installing in some managed notebooks may require a restart of the kernel.

!pip install pygam
# !pip install scikit-learn
# !pip install pandas matplotlib seaborn

# Optional: if a 'fastsparse' package exists (research code), install it here.

#!pip install fastsparse  # (hypothetical)

# Core imports
import os
import io
import math
import joblib
import warnings
from datetime import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, LassoCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

# Try to import fastsparse if available; otherwise use pygam
try:
    import fastsparse
    HAVE_FASTSPARSE = True
except Exception:
    HAVE_FASTSPARSE = False

from pygam import LinearGAM, LogisticGAM, s, f

warnings.filterwarnings('ignore')

print("Environment ready. HAVE_FASTSPARSE=", HAVE_FASTSPARSE)

# ---------------------------
# CELL 2 — Load dataset
# ---------------------------

# Replace this path with your CSV path or use file upload in Jupyter.
CSV_PATH = 'Chiller_PdM_dataset.csv'  # change if needed

if not os.path.exists(CSV_PATH):
    print(f"WARNING: {CSV_PATH} not found. Please upload your CSV and set CSV_PATH accordingly.")
else:
    df = pd.read_csv(CSV_PATH)
    print("Loaded dataset:", CSV_PATH)
    display(df.head())
    print(df.info())

# ---------------------------
# CELL 3 — Quick preprocessing & target detection
# ---------------------------

# Basic helpers

def detect_task_and_target(df):
    # assume last column is the target
    target = df.columns[-1]
    y = df[target]
    if pd.api.types.is_numeric_dtype(y) and set(np.unique(y)).issubset({0,1}):
        task = 'classification'
    elif pd.api.types.is_numeric_dtype(y):
        task = 'regression'
    else:
        task = 'classification'  # fallback: will encode
    return task, target

if 'df' in globals():
    task, target = detect_task_and_target(df)
    print('Detected task:', task, '| target column:', target)

# Drop identifier-like columns (strings)
if 'df' in globals():
    nonnum = df.select_dtypes(include=['object']).columns.tolist()
    if len(nonnum) > 0:
        print('Dropping non-numeric columns (identifiers):', nonnum)
        df = df.drop(columns=nonnum)

# Show data distribution
if 'df' in globals():
    display(df.describe())
    if task=='classification':
        print('Class balance:')
        print(df[target].value_counts(normalize=True))

# ---------------------------
# CELL 4 — Feature engineering: lags + rolling stats
# ---------------------------

def create_lag_features(df, cols, n_lags=5):
    df2 = df.copy()
    for c in cols:
        for lag in range(1, n_lags+1):
            df2[f"{c}_lag{lag}"] = df2[c].shift(lag)
    return df2

def create_rolling_features(df, cols, windows=[3,5,10]):
    df2 = df.copy()
    for c in cols:
        for w in windows:
            df2[f"{c}_rollmean_{w}"] = df2[c].rolling(window=w, min_periods=1).mean()
            df2[f"{c}_rollstd_{w}"] = df2[c].rolling(window=w, min_periods=1).std().fillna(0)
    return df2

if 'df' in globals():
    sensor_cols = [c for c in df.columns if c != target]
    print('Sensor columns:', sensor_cols)
    df_fe = create_lag_features(df, sensor_cols, n_lags=5)
    df_fe = create_rolling_features(df_fe, sensor_cols, windows=[3,5])
    df_fe = df_fe.dropna().reset_index(drop=True)
    print('After FE shape:', df_fe.shape)
    display(df_fe.head())

# ---------------------------
# CELL 5 — Train/test split
# ---------------------------

if 'df_fe' in globals():
    X = df_fe.drop(columns=[target])
    y = df_fe[target]

    # Keep numeric features only
    X = X.select_dtypes(include=[np.number])

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.4, random_state=42, stratify=(y if task=='classification' else None)
    )
    print('Train shape:', X_train.shape, 'Test shape:', X_test.shape)

# ---------------------------
# CELL 6 — Sparse feature selection (L1) or tree-based
# ---------------------------

# We'll run a simple L1-based feature selection using Lasso (for regression) or
# LogisticRegression with L1 penalty for classification.

selected_features = None

if 'X_train' in globals():
    print('Running sparse feature selection...')
    # scale features
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X_train)

    if task == 'regression':
        # LassoCV
        lasso = LassoCV(cv=5, random_state=42, n_jobs=-1).fit(Xs, y_train)
        coef = pd.Series(lasso.coef_, index=X_train.columns)
        selected_features = coef[coef!=0].index.tolist()
    else:
        # Logistic with L1
        clf = LogisticRegression(penalty='l1', solver='saga', max_iter=5000, C=1.0)
        clf.fit(Xs, y_train)
        coef = pd.Series(clf.coef_.ravel(), index=X_train.columns)
        selected_features = coef[coef!=0].index.tolist()

    if len(selected_features) == 0:
        # fallback: take top k from tree-based importance
        print('L1 selected 0 features — falling back to RandomForest importances')
        rf = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train)
        imp = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False)
        selected_features = imp.head(10).index.tolist()

    print('Selected features ({}):'.format(len(selected_features)), selected_features)

# ---------------------------
# CELL 7 — Fit GAM on selected features
# ---------------------------

if 'selected_features' in globals() and selected_features is not None and len(selected_features)>0:
    Xg_train = X_train[selected_features]
    Xg_test = X_test[selected_features]

    print('Fitting GAM on {} features'.format(len(selected_features)))

    # Build terms: one spline per feature
    terms = s(0) # Initialize with the first term
    for i in range(1, len(selected_features)):
        terms += s(i) # Add subsequent terms

    if task == 'regression':
        gam = LinearGAM(terms).gridsearch(Xg_train.values, y_train.values)
    else:
        gam = LogisticGAM(terms).gridsearch(Xg_train.values, y_train.values)

    print('GAM fitted. Summary:')
    print(gam.summary())

# ---------------------------
# CELL 8 — Evaluate model
# ---------------------------

if 'gam' in globals():
    if task=='classification':
        preds = gam.predict(Xg_test.values)
        raw_probs = gam.predict_proba(Xg_test.values)
        if raw_probs.ndim == 2:
            probs = raw_probs[:, 1]
        else:
            probs = raw_probs # Assume 1D array is probability of positive class
        print(classification_report(y_test, preds))
        print('ROC AUC:', roc_auc_score(y_test, probs))
    else:
        preds = gam.predict(Xg_test.values)
        from sklearn.metrics import mean_squared_error
        print('RMSE:', math.sqrt(mean_squared_error(y_test, preds)))

# ---------------------------
# CELL 9 — Interpretability: Plots & Partial Dependence
# ---------------------------

if 'gam' in globals():
    # term plots
    fig, axes = plt.subplots(math.ceil(len(selected_features)/2), 2, figsize=(12, 4*math.ceil(len(selected_features)/2)))
    axes = axes.flatten()
    for i, feat in enumerate(selected_features):
        XX = gam.generate_X_grid(term=i)
        ax = axes[i]
        pdep, confi = gam.partial_dependence(term=i, X=XX, width=0.95)
        ax.plot(XX[:, i], pdep)
        # Check if confi is 2D (lower/upper bounds) or 1D (symmetric interval)
        if confi.ndim == 2:
            ax.fill_between(XX[:, i], confi[:, 0], confi[:, 1], alpha=0.2)
        else:
            ax.fill_between(XX[:, i], pdep - confi, pdep + confi, alpha=0.2)
        ax.set_title(f'Partial dependence: {feat}')
        ax.set_xlabel(feat)
    plt.tight_layout()
    plt.show()

    # Feature importance via absolute term strengths
    term_stats = pd.DataFrame({'feature': selected_features, 'abs_strength': np.abs(gam.coef_.ravel()[:len(selected_features)])})
    term_stats = term_stats.sort_values('abs_strength', ascending=False)
    display(term_stats)

# ---------------------------
# CELL 10 — Save model + artifacts
# ---------------------------

if 'gam' in globals():
    ts = datetime.now().strftime('%Y%m%d_%H%M%S')
    model_file = f'gam_pdm_model_{ts}.joblib'
    joblib.dump({'model': gam, 'scaler': scaler, 'features': selected_features}, model_file)
    print('Saved model to', model_file)

# ---------------------------
# CELL 11 — Diagram & explanation
# ---------------------------

from IPython.display import Markdown, display

explan_text = '''
## Diagram & Explanation — FASTSPARSE-like GAM PdM Pipeline

1. Input: sensor time-series (temperature, vibration, pressure, ...)
2. Feature engineering: create lag features and rolling statistics to capture temporal context
3. Sparse feature selection: L1-based selection or tree-based importance to reduce dimensionality
4. GAM model: fit spline-based terms for each selected feature (interpretable smooth functions)
5. Interpretation: partial dependence plots show how each sensor influences failure risk
6. Deployment: save the trained GAM + scaler + feature list, use in production for inference

This notebook uses pyGAM as a robust, open-source fallback. If a FastSparse implementation is available,
replace the GAM-fit cell with the FastSparse training call (keeps same inputs: X_train, y_train)
'''

display(Markdown(explan_text))

# End of notebook
```
